```python
# EDA y calidad de datos
import pandas as pd
from fed_fraud.data_utils import basic_cleaning, data_quality_report, validate_dataset

# Cargar
df = pd.read_csv('../data/fraud_data.csv')
print('Raw rows, cols:', df.shape)

# Limpieza básica
df_clean = basic_cleaning(df, label_col='label')
print('After basic cleaning rows, cols:', df_clean.shape)

# Reporte de calidad
report = data_quality_report(df_clean, label_col='label')
import json
print(json.dumps(report, indent=2))

# Validación de requisitos (3 bancos, mínimo 5000 filas por banco, al menos 6 features)
val = validate_dataset(df_clean, label_col='label', n_clients=3, min_per_client=5000, min_features=6)
print('Validation result:')
print(val)
```
